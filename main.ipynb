```python
import pandas as pd
from requests import get
from urllib.parse import quote
API_KEY = '7789e585a72c1c95ab44a555ccc2b4e4'
```

```python
df = pd.read_csv('datasets/songs.csv').drop(columns=['index','acous', 'spch', 'pop'])
```

```python
def songUrl(artist, songname, key):
    return f"https://api.vagalume.com.br/search.php?art={quote(artist.encode('utf-8'))}&mus={quote(songname.encode('utf-8'))}&apikey={key}"

def requestLyrics(songs):
    lyrics = []
    for url in songs['url']:
        req = get(url)
        if req:
            res = req.json()
            # se existir na API && estiver escrita em inglês (limitações de encoding)
            if res['type'] in ['exact', 'aprox']:
                lyrics += [(url, res['mus'][0]['text'])]
            else: # temporario
                lyrics += [res['type']]
    return lyrics

def makeSongList(f):
    songs = pd.read_csv(f, ',')
    # add API URL column
    songs['url'] = pd.DataFrame([
        songUrl(row[0], row[1], API_KEY)
        for row in zip(songs['artist'], songs['title'])
    ], columns=['url'])

    return songs
```

```python
songsDir = 'datasets/songs.csv'
songs = makeSongList(songsDir)
lyrics = requestLyrics(songs)
songs.to_csv('songs.csv', index=False)
```
